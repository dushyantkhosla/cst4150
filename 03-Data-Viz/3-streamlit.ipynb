# Streamlit: Interactive Data Apps

Streamlit is a Python library that helps create web applications with ease. 
The application is written without any JS code.

The code is stored in `.py` file (called a Python script) such as `myapp.py` and executed from the command line using `streamlit run myapp.py`

We start with 

```python
import streamlit as st
```

and make use of the following functions

- `st.set_page_config`
- `st.title`
- `st.sidebar.header`
- `st.sidebar.selectbox`
- `st.header`
- `st.subheader`
- `st.write`

The app is separated into a **Sidebar** and a **Body** section
Generally, we allow users to input their choices in the sidebar, and load interactive visualizations in the body.
