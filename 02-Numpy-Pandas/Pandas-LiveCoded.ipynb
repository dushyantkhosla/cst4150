```python
mode_embarked = (df['Embarked'].value_counts(normalize=True).idxmax())
print(f"The mode of embarked is: {mode_embarked}")
```

```python
(df
 .assign(Age_filled=lambda fr: fr['Age'].fillna(average_age))
 .assign(Embarked_filled=lambda fr: fr['Embarked'].fillna(mode_embarked))
 .isnull()
 .astype(int)
 .mean()
)
```

```python
age_bins_arbit = (pd.cut(x=df['Age'].dropna(),
                         bins=[0, 2, 12, 18, 25, 45, 81],
                         labels=['infants', 'young kids', 'teens', 'young adults', 'middle age', 'elder']))
```

```python
df_2 = \
(df
  .dropna(subset='Age')
  .assign(age_bins=age_bins_arbit.values)
)
```
