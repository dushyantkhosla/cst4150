## What is Statistics?

In a nutshell, statistics is a set of *tools* and *procedures* that help us in

- summarizing or **explaining complex phenomena** using simple, reasonable probability **models** 
- Infer **underlying parameters** that may have generated the data
- generating **knowledge** about a population by analyzing **noisy** data from a sample (*inference*)
- **Make predictions** about unobserved data, or expected future observations.

## Types of questions that statistics can help us answer

- **Descriptive** to summarize a characteristic of a set of data
- **Exploratory** to identify patterns, trends, or relationships between variables
- **Inferential** to test a hypotheses
- **Predictive**  to forecast an unknown value or label for a case using a model build on a training set
- **Causal** to investigate whether changing one factor will change another factor in a population
- **Deterministic** to establish *how* the change in one factor results in change in another factor in a population - to determine the exact mechanism"

## Populations and Samples
---

* we can hardly ever measure metrics for a population
* instead, we need to work with samples

Statistics helps in drawing conclusions about the behavior of a population from a sample of data from that population
Measure the metrics we need, while quantifying uncertainty

When a sample (a set of observations) is drawn from a larger population, 
* the sample points have an empirical distribution that is discrete
* which provides information about the population distribution

<br>

A probability distribution is a mathematical description of the probabilities of events
It is common to distinguish between discrete and continuous variables

## Discrete Data

For discrete data, we specify a *probability mass function* $p$ that assigns a probability to each possible outcome:
* when throwing a fair die, each of the six digits 1 to 6 have an equal probability of appearing when a die is rolled $p(x=N)=1/6$
* when tossing a fair coin, $p(x=H)=p(x=T)=1/2$

The probability of an **event** is then defined to be the *sum of the probabilities of all outcomes that satisfy the event*
For example, rolling a die to get an even number: $P(x=Even) = P(x=2) + P(x=4) + P(x=6) = 1/2$

Well-known discrete probability distributions used in statistical modeling include 

* Poisson 
* Bernoulli 
* Binomial
* Negative binomial 
* Geometric 

<br>

## Continuous Data

We describe continuous variables by means of 
- the *cumulative distribution function*, which describes the probability that the variable is $P(X ≤ x)$
- the *probability density function* which describes the probability that $P(X = x)$

Well-known continuous probability distributions used in statistical modeling include

* Beta
* Gaussian
* Lognormal
* Exponential
* Gamma

If we know which distribution our data comes from, we can utilise their properties to answer questions such as ::


- What is the probability that a variable X takes on a value x?

    - For example, what is the probability that a diamond costs more than USD 5000
    - To calculate this, we use a *Cumulative Density Function, or CDF*
    - We can also calculate $P(X>x) = 1 - P(X <= x)$ 

<br>

- What is the probability that a variable X takes on a value in the range (x_1, x_2)
    -  $P(x_1 < X < x_2)$
    -  To calculate this, we use the difference of CDFs

A quick example using the Acceleration of cars data:

> Question: Calculate the probability that a car has an acceleration less than 10

````python
# Visualize 
df_cars.loc[:, 'Acceleration'].plot.hist(bins=20, figsize=(5, 3));
````

<big>

- We observe that this looks like a *Normal Distribution*
- To get the PDF or CDF of a normal distribution, we need its Mean (MU) and Standard Deviation (SIGMA)
- We can get these by **fitting** the data 

<br>

<br>

## The .fit() method
---

* Return estimates of shape (if applicable), location, and scale parameters from data
* The default estimation method is Maximum Likelihood Estimation (MLE)
* Method of Moments (MM) is also available.

<br>

````python
norm.fit?
````

````python
acc_mu, acc_sigma = norm.fit(df_cars.loc[:, 'Acceleration'].values)
````

<br>

Note that the `.fit()` method may return a different set of parameters depending on the distribution  
For example,

```python
from scipy.stats import gamma
shape, loc, scale = gamma.fit(mydata)
```

which would then be used as 

```python
x = np.linspace(data.min(), data.max(), data.count())
gamma.pdf(x, shape, loc, scale)
```

Okay, now back to our problem:
<br>

````python
print(f"\nFor Acceleration of Cars, MU={acc_mu:.2f} and SIGMA={acc_sigma:.2f}")
print("Now, let's calculate the probability that a car has an acceleration less than 10\n")
````

````python
proba_acc_le_10 = norm.cdf(x=10, loc=acc_mu, scale=acc_sigma)
print(f"\nThe probability that a car exists with acceleration P(X<=10) is {proba_acc_le_10:.2%}.\nWe can also verify this against the histogram\n")
````

````python
xs = np.linspace(start=df_cars['Acceleration'].min(), stop=df_cars['Acceleration'].max(), num=100)
````

````python
pd.Series(norm.pdf(xs, loc=acc_mu, scale=acc_sigma)).plot(figsize=(3, 2));
````

````python
ax = pd.Series(norm.cdf(xs, loc=acc_mu, scale=acc_sigma)).plot(figsize=(3, 2))
ax.vlines(x=10, ymin=0, ymax=0.2, linestyle='--', color='red');
````

````python
ax = pd.Series(norm.sf(xs, loc=acc_mu, scale=acc_sigma)).plot(figsize=(3, 2))
ax.vlines(x=10, ymin=0, ymax=0.9, linestyle='--', color='red');
````

## Methods of Statistical Distributions - `rvs, pdf, cdf, sf, stats` 
---


All functions concerning probability distributions are contained in `scipy.stats`
Overall, there's an exhaustive list of

* 81 continuous distribution families
    * examples include Beta, Cauchy, Chi-squared, Gamma, Weibull, Exponential, Pareto, Logistic, Gaussian
* 12 discrete distribution families
    * such as Bernoulli, Binomial, Poisson

For each of these, we have many crucial methods, such as:

* Generate random variates `.rvs()`
* Probability Density Function as `.pdf()`
* Cumulative Distribution Function as `.cdf()` and Survival Function as `.sf()
`* Summary metrics (Mean, Variance, Skewness, Kurtosis) as `.stats`()

There's a simple syntax to access these: `scipy.stats.<distribution family>.<function>`

<br>


### Probability Density Function, $PDF$
---

* The PDF, or density $PDF(x)$ of a continuous random variable, is a function that describes the probability for a random variable $X$ to take on a given value $x$.
* This function helps us answer questions such as: *What is the chance that a man is between 160 and 165 cm tall?*
* The integral over the PDF between $x=a$ and $x=b$ gives the probability of that $x$ lies in the range $(a, b)$

<img src='assets/pdf.png' width=300 height=300>

<br>

### Cumulative Density Function, $CDF$
---

* The CDF of a continuous random variable, is the function that describes the probability for a random variable $X$ to take on value less than $x$.
* This function helps us answer questions suchs as: *What is the chance that a man is less than 165 cm tall?*
* Thus, the probability of that $x$ lies in the range $(a, b)$ will be given by $CDF(x=b) - CDF(x=a)$

## Location and Scale

<big>

* Here, we generate data from 3 normal distributions, each with a difference loc and scale value.
* Note how the loc controls position along the X-axis and scale controls the spread of the bell-curve.

> The green histogram corresponds to `loc=-5` and `scale=0.3` which is why it's located to the far left and has a narrow bell-shape.

````python
pd.Series(norm.rvs(loc=0, scale=1, size=1000)).plot.hist(figsize=(15, 4))
pd.Series(norm.rvs(loc=-5, scale=0.3, size=1000)).plot.hist()
pd.Series(norm.rvs(loc=6, scale=0.8, size=1000)).plot.hist();
````

# Central Limit Theorem
---

> Sampling distribution of means always follows a normal distribution


````python
df_diamonds['price'].mean()
````

````python
df_diamonds['price'].plot.hist(figsize=(3, 3))
````

### Sample Mean

````python
df_diamonds['price'].sample(frac=0.1, replace=True).mean()
````

````python
def get_sample_mean_price_diamonds():
    return df_diamonds['price'].sample(frac=0.1, replace=True).mean()
````

````python
srs_sampling_distribution = pd.Series([get_sample_mean_price_diamonds() for x in range(10000)])
````

````python
ax = srs_sampling_distribution.plot.hist(figsize=(3, 3), title=srs_sampling_distribution.mean())
ax.vlines(x=srs_sampling_distribution.mean(), ymin=0, ymax=3100, linestyle='--', color='green', alpha=0.4);
````
